### Load Packages

```python
pollution_data = pd.read_csv('station_hour.csv')
pollution_data.head()
```

### Preprocessing

```python
pollution_data.info()
```

```python
# Sort the values
pollution_data = pollution_data.sort_values(by='Datetime', ascending=True)
pollution_data
```

```python
# Drop null values
# pollution_data = pollution_data.dropna()
pollution_data = pollution_data.fillna(0)
pollution_data
```

```python
# Convert Datetime to date
pollution_data['Datetime'] = pd.to_datetime(pollution_data['Datetime'])
pollution_data['Date'] = pollution_data['Datetime'].dt.date
pollution_data['Date'] = pd.to_datetime(pollution_data['Date'])
pollution_data.info()
```

### Question 1

Use line plot to visualize the daily average of ALL pollutant measurements in India.  Observe proper axes labelling, with corresponding units and titles.


```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3',
              'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

index = pollution_data.set_index('Datetime')
daily_avg = index[pollutants].resample('D').mean()

plt.figure(figsize=(14, 8))

for pollutant in pollutants:
    plt.plot(daily_avg.index, daily_avg[pollutant], label=pollutant)

plt.title("Daily Average of Air Pollutants in India")
plt.xlabel("Date")
plt.ylabel("Concentration (various units)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3',
              'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

daily_avg = pollution_data.groupby('Date')[pollutants].mean(numeric_only=True)

plt.figure(figsize=(14, 8))

for pollutant in pollutants:
    plt.plot(daily_avg.index, daily_avg[pollutant], label=pollutant)

plt.title("Daily Average of Air Pollutants in India")
plt.xlabel("Date")
plt.ylabel("Concentration")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# for pollutant in pollutants:
#     fig, ax = plt.subplots(figsize=(10, 8))

#     ax.plot(daily_avg.index, daily_avg[pollutant], label=pollutant)
#     ax.set_xlabel('Date')
#     ax.set_ylabel('Concentration')
#     ax.set_title(f'Daily Average of Air Pollutants in India')
#     ax.legend()
#     plt.show()
```

### Question 4

What is the comparative difference in pollution levels between the lockdown period in  2020 and the same calendar months in 2019? Do this for each pollutant and express the  difference in terms of percent change. Use the daily average.

```python
# Lockdown from March to May
lockdown_2020 = pollution_data[(pollution_data['Datetime'] >= '2020-03-01') & (pollution_data['Datetime'] <= '2020-05-31')]
lockdown_2019 = pollution_data[(pollution_data['Datetime'] >= '2019-03-01') & (pollution_data['Datetime'] <= '2019-05-31')]

daily_avg_2020 = lockdown_2020.groupby('Date')[pollutants].mean()
daily_avg_2019 = lockdown_2019.groupby('Date')[pollutants].mean()

mean_2020 = daily_avg_2020.mean()
mean_2019 = daily_avg_2019.mean()

percent_change = ((mean_2020 - mean_2019) / mean_2019) * 100
percent_change = percent_change.round(2)
print("Percent change in pollutant levels (Lockdown 2020 vs. 2019):")
print(percent_change.sort_values())
```
